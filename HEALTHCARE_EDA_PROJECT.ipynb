```python
#1. How many male and female patients are in the dataset?
count = health['Gender'].value_counts()
count.columns=count.sum()
count
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
plt.figure(figsize=(4,6))
sns.barplot(x=count.index, y=count.values,palette='colorblind')
plt.ylabel('count')
plt.show()
```

```python
# 2.How many patients are in each medical condition according gender?
Gender_count =health.groupby(['Medical Condition', 'Gender']).size().unstack(fill_value=0)
print(Gender_count)
```

```python
# 5.How many patients got dicharged on each date?
discharge = health['Discharge Date'].value_counts().reset_index()
discharge.columns= ['Discharge Date', 'Patient Count']
print(discharge)
```

```python
discharge['Discharge Date'] = pd.to_datetime(discharge['Discharge Date']) 
discharge = discharge.sort_values('Discharge Date')
```

```python
discharge['Month-Year'] = discharge['Discharge Date'].dt.to_period('M')  
monthly_discharge = discharge.groupby('Month-Year').agg({'Patient Count': 'sum'}).reset_index()
```

```python
plt.figure(figsize=(12, 6))
sns.barplot(data=monthly_discharge, x='Month-Year', y='Patient Count',palette='viridis')
plt.title('Number of Patients Discharged by Month')
plt.xlabel('Month-Year')
plt.ylabel('Patient Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

From this we can understand that the number of patients discharged over time. It helps identify trends, such as certain months with higher discharge counts. By regularly monitoring discharge trends, the healthcare facility can better anticipate needs, improve patient flow, and enhance overall service delivery.

```python
# 13.What is the average length of hospital stay for each medical condition?
health[['Date of Admission','Discharge Date']]
```

```python

health['Date of Admission'] = pd.to_datetime(health['Date of Admission'])
health['Discharge Date'] = pd.to_datetime(health['Discharge Date'])

health['duration']=(health['Discharge Date']-health['Date of Admission']).dt.days
health['duration']
```

```python
duration_time=health.groupby('Medical Condition')['duration'].median().reset_index()
duration_time.columns=['Medical Condition','Days']
duration_time
```

```python
health['Date of Admission'] = pd.to_datetime(health['Date of Admission'])
health['Discharge Date'] = pd.to_datetime(health['Discharge Date'])

health['Length of Stay'] = (health['Discharge Date'] - health['Date of Admission']).dt.days

health['Admission Year'] = health['Date of Admission'].dt.year

avg_lenofstay = health.groupby(['Medical Condition', 'Admission Year'])['Length of Stay'].mean().reset_index()

plt.figure(figsize=(12, 6))  
sns.barplot(x='Admission Year', y='Length of Stay', hue='Medical Condition', data=avg_lenofstay,palette='colorblind')

plt.xticks(rotation=45)

plt.xlabel('Year of Admission')
plt.ylabel('Average Length of Stay (Days)')
plt.title('Average Length of Stay for Each Medical Condition by Year')

plt.tight_layout()
plt.show()
```

It highlights how the average stay may vary over time for different conditions. Some conditions may have longer stays in certain years, potentially reflecting changes in treatment protocols, patient demographics, or hospital resource availability. The analysis helps identify trends and informs healthcare planning.

```python
# 14.How many patients of each age are there for each medical condition?
age_count=health.groupby('Medical Condition')['Age'].value_counts().reset_index().sort_values(by='Age',ascending=False)
age_count
```

```python
bins = [0, 18, 35, 50, 70, 100]  
labels = ['0-18', '19-35', '36-50', '51-70', '71+']  

health['Age Group'] = pd.cut(health['Age'], bins=bins, labels=labels, right=True)

age_group_counts = health.groupby(['Medical Condition', 'Age Group']).size().reset_index(name='Count')

age_group_counts = age_group_counts.sort_values(by=['Age Group'], ascending=True)

sns.barplot(x='Age Group', y='Count', hue='Medical Condition', data=age_group_counts)

plt.xticks(rotation=45)

plt.xlabel('Age Group')
plt.ylabel('Number of Patients')
plt.title('Number of Patients by Age Group for Each Medical Condition')

plt.tight_layout()
plt.show()
```

Younger age groups tend to have more patients with conditions like Asthma and Infectious diseases.
Middle-aged and older adults are more likely to have chronic conditions such as Hypertension, Diabetes, and cancer.
This highlights the increasing prevalence of certain health conditions with age, emphasizing the need for age-specific healthcare approaches.

```python
# 16.what is the total billing amount for each insurance provider?
insurance_billing=health.groupby('Insurance Provider')['Billing Amount'].sum().sort_values(ascending=False).reset_index()
insurance_billing
```

From this we could understand that Cigna has the highest total billing amount compared to other insurance providers.This might mean that more people with Cigna insurance are getting treatment at the hospital, or the medical care they need is more expensive.It could also mean that people covered by Cigna have more serious health conditions that require costly treatments.
